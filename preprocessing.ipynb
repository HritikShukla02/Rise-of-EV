```python
def preprocess_by_month(df, year):
    # Drop first two rows
    if 'JAN' in df.iloc[2].values:
        df = df.iloc[2:, 1:]
    else:
        df = df.iloc[3:, 1:]

    # Extract column names from the first remaining row
    cols = df.iloc[0].astype(str).str.strip().values
    cols[0] = 'STATE'
    cols[-1] = 'TOTAL'
    df.columns = cols

    # Drop the header row from data
    df = df.iloc[1:].reset_index(drop=True)

    # Add year column
    df['year'] = year

    # Clean category names
    # df['Fuel_Type'] = df['vehicle_category'].str.strip()
    # df['vehicle_category'] = df['vehicle_category'].map({
    #     'FOUR WHEELER (INVALID CARRIAGE)': '4WIC',
    #     'LIGHT MOTOR VEHICLE': 'LMV',
    #     'MEDIUM MOTOR VEHICLE': 'MMV',
    #     'HEAVY MOTOR VEHICLE': 'HMV'
    # })

    # Reorder columns: 'year' first
    priority = ['year']
    rest = [col for col in df.columns if col not in priority]
    # print(df.head())
    df = df.reindex(columns=priority + rest)

    # Filter valid categories
    # df = df[df['vehicle_category'].isin(['4WIC', 'LMV', 'MMV', 'HMV'])]

    return df
```

```python
d = preprocess_by_month(data, 2024)
d
```

```python
data_by_months = pd.DataFrame()
for i in range(2002,2026):
    df = pd.read_excel(f'./data/State_ev_data/{i}.xlsx', header=None)
    print(i)
    pre_processed = preprocess_by_month(df, i)
    data_by_months = pd.concat((data_by_months, pre_processed), ignore_index=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
sns.set_style()
```

```python
raw_data = pd.DataFrame(columns=['YEAR', 'STATE', 'Fuel_Type', 'Registrations'])
for i in range(2002, 2026):
    df = pd.read_excel(F'./data/State_data/{i}.xlsx')
    data = df.iloc[2:, 1:]
    cols = list(data.iloc[0, 1:])
    cols.insert(0, 'STATE')
    cols.pop(-1)
    # print(cols)

    cols.append('TOTAL')
    # print(cols)
    data.columns = cols
    data = data.iloc[1:, :]

    data.drop(columns='TOTAL', inplace=True)
    data.index = range(data.shape[0])
    # print(data.index)
    data['YEAR'] = i
    melt = data.melt(id_vars=['YEAR', 'STATE'],var_name='Fuel_Type', value_name='Registrations')   
    raw_data = pd.concat([raw_data, melt])
# data
```

```python
raw_data.reset_index(drop=True, inplace=True)
```
